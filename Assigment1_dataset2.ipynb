```python
import pandas as pd
data = pd.read_csv('/content/Phone dataset.csv')
total_duration = data['network_type'].apply(lambda x: x).min()
print(total_duration)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Phone dataset.csv')
plt.hist(data['network'],bins=15)
plt.title('HISTOGRAM')
plt.xlabel('Network')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Phone dataset.csv')
plt.scatter(data['date'], data['network_type'])
plt.title('SCATTER PLOT')
plt.xlabel('Date')
plt.ylabel('Network_type')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Phone dataset.csv')
plt.figure(figsize=(7,4))
sns.boxplot(x=data['network'])
plt.title('BOX PLOT')
plt.xlabel('Network')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Phone dataset.csv')
plt.figure(figsize=(8,5))
sns.lineplot(x=data['network_type'], y=data['month'])
plt.title('LINE PLOT')
plt.xlabel('Network_type')
plt.ylabel('Month')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Phone dataset.csv')
plt.figure(figsize=(6,8))
plt.pie(data['network'].value_counts(), labels = data['network'].unique(), autopct='%1.1f%%')
plt.title('PIE CHART')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Phone dataset.csv')
plt.figure(figsize=(5,4))
plt.bar(data['item'].head(15), data['index'].head(15))
plt.title('BAR PLOT')
plt.xlabel('Item')
plt.ylabel('Index')
plt.show()
```
