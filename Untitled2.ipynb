```python
import pandas as pd
data = pd.read_csv("/content/employees.csv")
data['EMPLOYEE_ID   '] = data['MANAGER_ID'].map({'DEPARTMENT_ID': 'DEPARTMENT_ID'})
print(data)
```

```python
import pandas as pd
data = pd.read_csv("/content/employees.csv")
filtered_data = data[data['SALARY'] > 10000]
print(filtered_data)
```

```python
import pandas as pd
from functools import reduce
data = pd.read_csv("/content/employees.csv")
total_HIRE_DATE_by_LAST_NAME=data.groupby('LAST_NAME')['HIRE_DATE'].apply(lambda x: reduce(lambda x, y: x + y, x))
print(total_HIRE_DATE_by_LAST_NAME)
```

```python
import pandas as pd
data = pd.read_csv("/content/employees.csv")
data['Salary_range'] = data['SALARY'].apply(lambda x: 'low' if x < 15000 else 'medium' if x > 8000 else 'high')
print(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/employees.csv")
plt.bar(data['DEPARTMENT_ID'].value_counts().index, data['DEPARTMENT_ID'].value_counts())
plt.title("BAR CHART")
plt.xlabel("SALARY")
plt.ylabel("count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/employees.csv")
plt.scatter(data['MANAGER_ID'], data['DEPARTMENT_ID'])
plt.title("MANAGER_ID VS.DEPARTMENT_ID")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/employees.csv")
plt.hist(data['SALARY'], bins=10)
plt.title("HISTOGRAM")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("/content/employees.csv")
sns.boxplot(x='DEPARTMENT_ID', y='EMPLOYEE_ID', data=data)
plt.title("BOX PLOT")
plt.show()
```
