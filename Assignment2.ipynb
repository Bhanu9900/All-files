```python
import pandas as pd
from scipy import stats
df = pd.read_csv("/content/employees.csv")
contingency_table = pd.crosstab(df["JOB_ID"], df["DEPARTMENT_ID"])
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)
print("Chi-Square statistic:", chi2_stat)
print("P-value:", p_val)
```

```python
import pandas as pd
from scipy import stats
df = pd.read_csv("/content/employees.csv")
t_stat, p_val = stats.ttest_ind(df[df["JOB_ID"] == "AC_ACCOUNT"]["SALARY"], df[df["JOB_ID"] == "IT_PROG"]["SALARY"])
print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
import pandas as pd
from scipy import stats
df = pd.read_csv("/content/employees.csv")
f_stat, p_val = stats.f_oneway(df[df["JOB_ID"] == "AD_VP"]["SALARY"], df[df["JOB_ID"] == "ST_MAN"]["SALARY"])
print("F-statistic:", f_stat)
print("P-value:", p_val)
```

```python
import numpy as np
import pandas as pd
data=pd.read_csv("/content/employees.csv")
X=data['EMPLOYEE_ID']
Y=data['SALARY']
def calculate_covariance(X,Y):
  n=len(X)
  if n!=len(Y):
    raise ValueError("Datasets  X and Y must have the same length")
  mean_X=np.mean(X)
  mean_Y=np.mean(Y)
  covariance=sum((X[i]-mean_X)*(Y[i]-mean_Y)for i in range(n))/(n-1)
  return covariance
covariance=calculate_covariance(X,Y)
print("covariance:",covariance)
```

```python
import numpy as np
import pandas as pd
data=pd.read_csv("/content/employees.csv")
X=data['EMPLOYEE_ID']
Y=data['SALARY']
def calculate_correlation(X, Y):
  covariance = calculate_covariance(X, Y)
  std_dev_X = np.std(X, ddof=1)
  std_dev_Y = np.std(Y, ddof=2)
  correlation = covariance / (std_dev_X * std_dev_Y)
  return correlation
correlation = calculate_correlation(X, Y)
print("correlation:",correlation)
```
