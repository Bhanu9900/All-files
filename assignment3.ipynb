```python
#z_scores outluietr detection
import pandas as pd
df = pd.read_csv('/content/GlobalWeatherRepository.csv')
latitude_column = 'latitude'
country_column = 'humidity'
Q1 = df[latitude_column].quantile(0.99)
Q3 = df[country_column].quantile(0.80)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df[latitude_column] >= lower_bound) & (df[country_column] <= upper_bound)]
print(df_no_outliers)
```

```python
#Bivariate outlier Detection
import pandas as pd
df['z_score']=((df['longitude'] - df['longitude'].mean()) / df['longitude'].std())**2 + \
                 ((df['wind_degree'] - df['wind_degree'].mean()) / df['wind_degree'].std())**2
outliers = df[df['z_score'] > 2]
print("Bivariate Outliers:")
print(outliers)
```

```python
#Timeserise outliers detection
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/content/GlobalWeatherRepository.csv')
data['wind_degree'] = pd.to_datetime(data['wind_degree'])
# Calculate rolling mean and standard deviation
rolling_mean = data['temperature_celsius'].rolling(window=30).mean()
rolling_std = data['temperature_celsius'].rolling(window=30).std()
outliers = np.abs(data['temperature_celsius'] - rolling_mean) > 2 * rolling_std
print("Time Series Outliers:")
print(data[outliers])
data=data[~outliers]
plt.plot(data.index, data['temperature_celsius'])
plt.xlabel('wind_degree')
plt.ylabel('temperature_celsius')
plt.title('Time Series Outliers Deletion')
plt.show()
```

```python
#Principla component analysis
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
df=pd.read_csv("/content/GlobalWeatherRepository.csv")
numeric_df = df.select_dtypes(include=['number'])
imputer = SimpleImputer(strategy='mean')
numeric_df_imputed = imputer.fit_transform(numeric_df)
pca=PCA(n_components=2)
reduced_data=pca.fit_transform(numeric_df_imputed)
reduced_df=pd.DataFrame(reduced_data,columns=['PC1','PC2'])
print(reduced_df)
plt.pie(np.abs(reduced_df['PC1']), labels=reduced_df.index)
plt.xlabel('wind_degree')
plt.ylabel('longitude')
plt.title('PCA')
plt.show()
```

```python
#Linear Discriminant Analysis(LDA)
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
lda=LinearDiscriminantAnalysis(n_components=2)
x_lda=lda.fit_transform(x,y)
plt.scatter(x_lda[:,0],x_lda[:,1],c=y,cmap='viridis')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('Linear Discriminant Analysis')
plt.show()
```

```python
#Discretization
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/GlobalWeatherRepository.csv")
df['humidity']=pd.qcut(df['humidity'],q=3,labels=['low','medium','high'])
print(df)
humidity = df['humidity'].value_counts()
plt.bar(humidity.index, humidity.values)
plt.xlabel('humidity')
plt.ylabel('Count')
plt.title('Distribution of Humidity Categories')
plt.show()
```

```python
#t-Distributed Stochastic Neighbor Embedding(t-SNE)
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
df=pd.read_csv("/content/GlobalWeatherRepository.csv")
numeric_df = df.select_dtypes(include=['number'])
imputer = SimpleImputer(strategy='mean')
numeric_df_imputed = imputer.fit_transform(numeric_df)
tsne = TSNE(n_components=2)
reduced_data=tsne.fit_transform(numeric_df_imputed)
reduced_df=pd.DataFrame(reduced_data,columns=['Dimension 1','Dimension 2'])
print(reduced_df)
plt.scatter(reduced_df['Dimension 1'],reduced_df['Dimension 2'])
plt.xlabel('latitude')
plt.ylabel('temperature_celsius')
plt.title('t-SNE Visualization')
plt.show()
```
